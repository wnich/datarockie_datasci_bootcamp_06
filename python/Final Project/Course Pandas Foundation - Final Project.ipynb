```python
# TODO 07 - which top 5 States have the least total sales between 15 April 2019 - 31 December 2019
df[df['Order Date'].between('2019-4-15','2019-12-31')].groupby('State')['Sales'].sum().nsmallest(5)
```

```python
# TODO 08 - what is the proportion of total sales (%) in West + Central in 2019 e.g. 25% 
# (total sales in west in 2019 + total sales in central in 2019) / total sales in 2019
#df[(df['Order Date'].dt.year == 2019) & (df.Region.isin(['West','Central']))].Sales.sum()/df[df['Order Date'].dt.year == 2019].Sales.sum()*100 

sales_cw_2019 = df[df['Order Date'].dt.year == 2019]
sales_cw_2019 = sales_cw_2019.query("Region in ('West','Central')")['Sales'].sum()
sales_2019 = df[df['Order Date'].dt.year == 2019].Sales.sum() #cant ['Sales']

percent_cw = (sales_cw_2019/sales_2019)*100
percent_cw
```

```python
# TODO 09 - find top 10 popular products in terms of number of orders vs. total sales during 2019-2020
#group by Product Name...Quantity count
#group by Product Name...sum of sales
popular_orders = df[df['Order Date'].between('2019','2020')].groupby('Product Name')['Sales'].agg(number_orders = 'count', total_sales ='sum').sort_values(['number_orders','total_sales'],ascending=False)
popular_orders.head(10)
```

```python
# TODO 10 -1 - plot at least 2 plots, any plot you think interesting :)
#Total profit each category in each year

total_profit_year = df['Order Date'].dt.year
total_profit = df.groupby( df['Order Date'].dt.year)['Profit'].sum()
#total_profit

total_profit_plot = total_profit.plot(kind='pie', y = 'Profit', autopct='%1.0f%%', startangle=60, title = 'Total profit of each category in each year')
```

```python
# TODO 10-2 
#5 most profitable Sub-Category in 2020
most_profit = df[df['Order Date'].dt.year == 2020].groupby('Sub-Category')['Profit'].sum().nlargest(5)
colors = ['darkorange', 'khaki', 'lightcoral', 'mediumpurple', 'lightskyblue']
most_profit_plot = most_profit.plot(kind='pie', y = 'Profit', autopct='%1.0f%%', startangle=60, colors = colors,  title='5 most profitable Sub-Categories in 2020')
```
